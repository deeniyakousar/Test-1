```python
# question 4
df = pd.DataFrame({'Travel Dates': ['2024-05-27 - 2025-06-02']})

# Split start and end dates
df[['Start Date', 'End Date']] = df['Travel Dates'].str.split(' - ', expand=True)

# Convert to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['End Date'] = pd.to_datetime(df['End Date'])

# Calculate duration in days
df['Trip Duration'] = (df['End Date'] - df['Start Date']).dt.days

print(df[['Travel Dates', 'Trip Duration']])
```
